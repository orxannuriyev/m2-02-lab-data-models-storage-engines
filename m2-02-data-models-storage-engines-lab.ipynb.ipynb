Show what the data would look like as a single denormalized table (join all three tables). Write a markdown cell explaining what redundancy appears and why the normalized version is preferable for updates.

***Normalization demo (Denormalized vs Normalized)***

**Denormalized table (joined view):** When we join `rentals` + `customers` + `movies` into one table, each rental row repeats customer and movie attributes.

**Redundancy that appears:**
- Customer fields (`name`, `email`, `city`, `signup_date`) are duplicated for every rental that the same customer makes.
- Movie fields (`title`, `genre`, `release_year`, `rating`) are duplicated for every rental of the same movie.

**Why the normalized version is better for updates:**
- In the normalized design, customer details live only in `customers` and movie details live only in `movies`.  
  If a customer changes email or a movie’s metadata needs correction, you update **one row** in the appropriate table.
- In a denormalized table, the same information may exist in many rows, so updates require changing **many rows**, which is slower and risks inconsistencies (some rows updated, others not).
- Keeping `rentals` as a fact table with only keys (`customer_id`, `movie_id`) plus rental-specific attributes (`rental_date`, `return_date`, `price`) reduces duplication and preserves data integrity.

# Task 2: Model the Same Data as Documents
Represent the same movie rental data using the document model.

## Step 1: Create a list of customer documents (Python dictionaries)
where each document contains the customer's information and a nested list of their rental history, with movie details embedded in each rental entry.

```python
customers.head()
```

```python
rentals.head()
```

```python
movies.head()
```

```python
# 1) rentals + movies join: hər rental-a movie detallarını embed etmək üçün
rentals_with_movies = rentals.merge(
    movies,
    on="movie_id",
    how="left",
    suffixes=("", "_movie")
)

# 2) Customer docs: hər customer üçün nested rental_history
customer_docs = []
for _, c in customers.iterrows():
    # həmin customer-a aid rental-ları seç
    r = rentals_with_movies[rentals_with_movies["customer_id"] == c["customer_id"]].copy()
    r = r.sort_values("rental_date")

    rental_history = []
    for _, row in r.iterrows():
        rental_history.append({
            "rental_id": int(row["rental_id"]),
            "rental_date": row["rental_date"],
            "return_date": row["return_date"],
            "price": float(row["price"]),
            "movie": {  # EMBEDDED movie details
                "movie_id": int(row["movie_id"]),
                "title": row["title"],
                "genre": row["genre"],
                "release_year": int(row["release_year"]),
                "rating": int(row["rating"]),
            }
        })

    customer_docs.append({
        "customer_id": int(c["customer_id"]),
        "name": c["name"],
        "email": c["email"],
        "city": c["city"],
        "signup_date": c["signup_date"],
        "rental_history": rental_history
    })

# Nəticəyə baxış (məs: 1-ci customer doc)
customer_docs[0]
```

Find all movies rented by a specific customer.


```python
# 1) Find all movies rented by a specific customer
customer_id = 1

movies_rented = [
    r["movie"]["title"]
    for doc in customer_docs
    if doc["customer_id"] == customer_id
    for r in doc["rental_history"]
]

movies_rented
```

Find the top 5 most-rented movies.


```python
# 2) Top 5 most-rented movies
from collections import Counter

movie_counts = Counter(
    r["movie"]["title"]
    for doc in customer_docs
    for r in doc["rental_history"]
)

top5_most_rented = movie_counts.most_common(5)
top5_most_rented  # list of (title, count)
```

Compute total revenue per genre.


```python
# 3) Total revenue per genre
from collections import defaultdict

revenue_by_genre = defaultdict(float)

for doc in customer_docs:
    for r in doc["rental_history"]:
        revenue_by_genre[r["movie"]["genre"]] += float(r["price"])

dict(revenue_by_genre)
```

Find customers who have never rented a movie rated below 3.


```python
# 4) Customers who have never rented a movie rated below 3
good_customers = [
    doc["name"]
    for doc in customer_docs
    if all(r["movie"]["rating"] >= 3 for r in doc["rental_history"])
]

good_customers
```

## Step 3: Write a markdown cell comparing the document queries to the relational queries. Which queries were easier to write? Which were harder? For which query pattern does the document model have a clear advantage?



### Comparison: Document Model vs Relational Model

In the relational model, the data is stored in separate tables (`customers`, `movies`, and `rentals`). To answer many questions, we must join these tables using operations like `merge`. This makes the structure normalized and avoids redundancy, but queries often require multiple steps.

In the document model, each customer document already contains their rental history with embedded movie information. Because related data is stored together, some queries become simpler since no joins are required.

**Easier in the document model:**
- Finding all movies rented by a specific customer was easier. The rentals are already nested inside the customer document, so we only iterate through that customer’s `rental_history`.

**Harder in the document model:**
- Queries that require aggregating data across all customers (such as finding the top 5 most-rented movies or total revenue per genre) were harder. We had to loop through all documents and rentals manually to compute counts or sums.

**Clear advantage of the document model:**
- The document model is particularly strong when accessing data that naturally belongs together, such as retrieving a customer's full rental history. This pattern benefits from having related information embedded in a single document, avoiding joins and making read operations straightforward.

# Task 3: Model Relationships as a Graph
Build a graph representation of the rental data to answer relationship-based questions.

## Step 1: Create a graph using Python dictionaries where:

Nodes represent customers, movies, and genres.
Edges represent relationships: rented (customer → movie), belongs_to (movie → genre), lives_in (customer → city).

```python
nodes = {
    "c001": {"type": "customer", "name": "Alice"},
    "c002": {"type": "customer", "name": "Bob"},

    "m001": {"type": "movie", "title": "Inception"},
    "m002": {"type": "movie", "title": "Titanic"},
    "m003": {"type": "movie", "title": "Avatar"},

    "sci_fi": {"type": "genre", "name": "Sci-Fi"},
    "romance": {"type": "genre", "name": "Romance"},
    "action": {"type": "genre", "name": "Action"},

    "london": {"type": "city", "name": "London"},
    "paris": {"type": "city", "name": "Paris"}
}
```

```python
edges = [
    ("c001", "rented", "m001"),
    ("c001", "rented", "m002"),
    ("c002", "rented", "m003"),

    ("m001", "belongs_to", "sci_fi"),
    ("m002", "belongs_to", "romance"),
    ("m003", "belongs_to", "action"),

    ("c001", "lives_in", "london"),
    ("c002", "lives_in", "paris")
]
```

## Step 2: Write traversal functions to answer:

"Which movies did customer X rent?" — one hop from customer through rented edges.
"Which genres does customer X prefer?" — two hops: customer → movies → genres, then count.
"Which customers share a genre preference with customer X?" — three hops: customer X → movies → genres → other customers who rented movies in the same genre.
"Recommend movies for customer X" — find movies in their preferred genres that they have not yet rented.


"Which movies did customer X rent?" — one hop from customer through rented edges

```python
def movies_rented_by_customer(customer_id):
    movies = []
    
    for src, rel, dst in edges:
        if src == customer_id and rel == "rented":
            movies.append(nodes[dst]["title"])
    
    return movies
```

```python
movies_rented_by_customer('c002')
```

"Which genres does customer X prefer?" — two hops: customer → movies → genres, then count.

```python
from collections import Counter

def preferred_genres(customer_id):
    
    genres = []
    
    for src, rel, movie in edges:
        if src == customer_id and rel == "rented":
            
            for m, r, genre in edges:
                if m == movie and r == "belongs_to":
                    genres.append(nodes[genre]["name"])
    
    return Counter(genres)
```

```python
preferred_genres('c001')
```

"Which customers share a genre preference with customer X?" — three hops: customer X → movies → genres → other customers who rented movies in the same genre.

```python
def similar_customers(customer_id):
    
    genres = set(preferred_genres(customer_id).keys())
    similar = set()
    
    for movie, rel, genre in edges:
        if rel == "belongs_to" and nodes[genre]["name"] in genres:
            
            for customer, r, m in edges:
                if r == "rented" and m == movie and customer != customer_id:
                    similar.add(nodes[customer]["name"])
    
    return list(similar)
```

```python
similar_customers('c001')
```

"Recommend movies for customer X" — find movies in their preferred genres that they have not yet rented.


```python
def recommend_movies(customer_id):
    
    rented = set()
    
    for src, rel, movie in edges:
        if src == customer_id and rel == "rented":
            rented.add(movie)
    
    genres = set(preferred_genres(customer_id).keys())
    
    recommendations = set()
    
    for movie, rel, genre in edges:
        if rel == "belongs_to" and nodes[genre]["name"] in genres:
            if movie not in rented:
                recommendations.add(nodes[movie]["title"])
    
    return list(recommendations)
```

```python
recommend_movies('c002')
```

## Step 3: Write a markdown cell explaining why the recommendation query (Task 3, question 4) 
would be difficult to express in SQL or with documents. What makes the graph model natural for this kind of traversal?



### Why the recommendation query is harder in relational or document models

The recommendation query requires following multiple relationships in sequence. In this task we traverse:  
**customer → movies they rented → genres of those movies → other movies in the same genres that the customer has not rented yet.**

In a relational database, this type of query requires several **JOIN operations** between tables such as `customers`, `rentals`, and `movies`. The query becomes complex because the same tables must be joined multiple times and additional conditions are needed to exclude movies already rented. As the number of relationships grows, the SQL query becomes longer and harder to maintain.

In the document model, the rental history is embedded inside customer documents. While retrieving a single customer's rentals is easy, answering a query that requires looking across many customers or genres requires iterating through many documents and manually aggregating results. This makes cross-document relationship queries more complicated and less efficient.

The **graph model** is natural for this problem because relationships are first-class elements of the data model. Traversing from one node to another through edges (customer → movie → genre → movie) is straightforward and matches the way the recommendation logic works. Instead of performing complex joins or scanning many documents, the graph simply follows connected edges. This makes multi-step relationship queries easier to express and more intuitive to implement.

## Step 4: Write a markdown cell explaining: 
Why would OLTP operations typically use row-major storage while OLAP operations use column-major storage? How does the access pattern of each workload type map to the memory layout that serves it best?



### Why OLTP uses row-major storage and OLAP uses column-major storage

OLTP (Online Transaction Processing) systems are designed for fast, frequent operations on individual records, such as inserting a transaction, updating a status, or retrieving a single row by its ID. These operations usually access **all attributes of one specific row** at once. In a **row-major storage layout**, all the columns of a row are stored together in memory. This means that when a single transaction record is accessed, the database can read the entire row efficiently in one contiguous block of memory. This layout minimizes disk and memory access for point queries and updates, which are common in OLTP workloads.

OLAP (Online Analytical Processing) systems, on the other hand, are optimized for analytical queries over large datasets. These queries typically perform **aggregations on specific columns**, such as calculating totals, averages, or counts across millions of records. In a **column-major storage layout**, values from the same column are stored together in contiguous memory. This allows the system to scan only the relevant columns needed for the calculation instead of reading every column in each row. As a result, column-major storage significantly improves performance for operations like grouping, filtering, and aggregating large datasets.

In summary, the choice of storage layout matches the **access pattern of the workload**. OLTP workloads access and modify **complete rows**, so row-major storage is more efficient. OLAP workloads analyze **specific columns across many rows**, so column-major storage enables faster scanning and more efficient analytical processing.

**Recommended data model:** Relational (column-oriented warehouse)  
**Processing paradigm:** OLAP  

A financial reporting system performs analytical queries that aggregate large volumes of transaction data to compute metrics such as monthly revenue by product category and region. These workloads involve scanning many records and performing operations like grouping, filtering, and summing values across columns. Relational data warehouses with column-oriented storage are well suited for this type of analytical processing because they allow efficient column scans and aggregations. Since the focus is on reporting and analysis rather than frequent transactional updates, an OLAP processing paradigm is the most appropriate choice.

**An IoT platform that receives sensor readings from 10,000 devices and needs to detect anomalies in real time.**

**Recommended data model:** Document  
**Processing paradigm:** OLTP  

An IoT platform receives continuous streams of sensor readings from thousands of devices, where each reading can naturally be represented as a document containing fields such as device_id, timestamp, and sensor values. The document model is flexible and works well with semi-structured data, allowing different devices to send slightly different data formats without requiring schema changes. The system must process incoming readings quickly and detect anomalies as soon as they arrive, which requires fast inserts and real-time queries. Because the workload involves high-frequency writes and immediate checks on individual records, it fits the characteristics of an OLTP processing paradigm.


**A content management system where each article has a different set of metadata fields.**

**Recommended data model:** Document  
**Processing paradigm:** OLTP  

A content management system often stores articles that may have different metadata fields such as author, tags, category, publication date, or custom attributes. Because the structure of this metadata can vary between articles, a document data model is well suited since it allows flexible and semi-structured data without requiring a fixed schema. Each article can be stored as a document with its own set of fields. The system mainly performs frequent operations such as creating, updating, and retrieving individual articles, which aligns with the characteristics of OLTP workloads.
